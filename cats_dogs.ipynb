```python
cat_images = os.listdir("archive (4)/PetImages/Cat")
dog_images = os.listdir("archive (4)/PetImages/Dog")

# training 80 | validation 10 | testinig 10
cat_images_training = cat_images[:10_000]
dog_images_training = dog_images[:10_000]
data_training = pd.DataFrame({"filename" : cat_images_training + dog_images_training, "dog?" : [False]*10_000 + [True]*10_000}).sample(frac=1).reset_index(drop=True)

cat_images_validation = cat_images[10_000:11_250]
dog_images_validation = dog_images[10_000:11_250]
data_validation = pd.DataFrame({"filename" : cat_images_validation + dog_images_validation, "dog?" : [False]*1_250 + [True]*1_250}).sample(frac=1).reset_index(drop=True)

cat_images_testing = cat_images[11_250:]
dog_images_testing = dog_images[11_250:]
data_testing = pd.DataFrame({"filename" : cat_images_testing + dog_images_testing, "dog?" : [False]*1_251 + [True]*1_251}).sample(frac=1).reset_index(drop=True)
```

```python

model = nn.Sequential(
    # Block 1
    nn.Conv2d(3, 32, kernel_size=3, padding=0),
    nn.BatchNorm2d(32),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2),
    nn.Dropout(p=0.2),

    # Block 2
    nn.Conv2d(32, 64, kernel_size=3, padding=0),
    nn.BatchNorm2d(64),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2),
    nn.Dropout(p=0.2),

    # Block 3
    nn.Conv2d(64, 128, kernel_size=3, padding=0),
    nn.BatchNorm2d(128),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2),
    nn.Dropout(p=0.2),

    # Block 4
    nn.Conv2d(128, 256, kernel_size=3, padding=0),
    nn.BatchNorm2d(256),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=2),
    nn.Dropout(p=0.2),

    # classification
    nn.Flatten(),
    nn.Linear(256*6*6, 512),
    nn.BatchNorm1d(512),
    nn.ReLU(inplace=True),
    nn.Dropout(p=0.5),
    nn.Linear(512, 2)
)
```
